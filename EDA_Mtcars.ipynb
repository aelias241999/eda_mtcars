###¿Autos pesados consumen más?

```python
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns

plt.scatter(df_work["wt"], df_work["mpg"])
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.title("Weigth vs MPG")
plt.show()

#Interactivo
fig=px.scatter(
    df_work,
    x="wt",
    y="mpg",
    color_discrete_sequence=["brown"]
)
fig.update_layout(
    plot_bgcolor="black", #fondo gráfico
    paper_bgcolor="black", #fondo exterior
    font=dict(color="white") #texto blanco
)
fig.show()

```
